```python
print(df['year'].unique())
```

```python
import re

def extract_oldest_year(value):
    # Extract all 4-digit numbers (years)
    years = re.findall(r'\b(18\d{2}|19\d{2}|20\d{2})\b', value)
    # Convert to integers and return the oldest one
    return min(map(int, years)) if years else None

# Apply function to the column
df['release_year'] = df['year'].astype(str).apply(extract_oldest_year)

print(df.release_year)
```

```python
#rearrange columns order
df = df.iloc[:, [0,9,2,3,4,6,5,7,8,1]] 
df.head()
```

```python
df.runtime.unique()
```

```python
#clean up runtime column
df['runtime'] = df['runtime'].str.replace('min', '', regex=False).str.strip()
df.runtime.unique()
```

```python
def clean_runtime(value):
    if pd.isna(value) or value == '':  # Handle NaN and empty strings
        return None  # Or use a default value like 0
    value = value.replace(',', '')  # Remove commas
    return int(value) if value.isdigit() else None  # Convert if it's numeric
    
df['runtime'] = df['runtime'].astype(str).apply(clean_runtime)
df.head()
```

```python
#change ratingValue to floats
def clean_ratingValue(value):
    if pd.isna(value) or value.strip() == '':  # Handle NaN and empty strings
        return None  # Or use a default like 0.0
    value = value.replace(',', '')  # Remove commas if present (e.g., '1,500.99')
    try:
        return float(value)  # Convert to float
    except ValueError:
        return None  # Handle non-numeric values like 'N/A'

# Apply function
df['ratingValue'] = df['ratingValue'].astype(str).apply(clean_ratingValue)

df.head()
```

```python
#change ratingCount to floats
def clean_ratingCount(value):
    if pd.isna(value) or value.strip() == '':  # Handle NaN and empty strings
        return None  # Or use a default like 0.0
    value = value.replace(',', '')  # Remove commas if present (e.g., '1,500.99')
    try:
        return float(value)  # Convert to float
    except ValueError:
        return None  # Handle non-numeric values like 'N/A'

# Apply function
df['ratingCount'] = df['ratingCount'].astype(str).apply(clean_ratingCount)

df.head()
```
