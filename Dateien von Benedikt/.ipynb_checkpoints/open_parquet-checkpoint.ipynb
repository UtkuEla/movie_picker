```python
import pandas as pd
```

```python
df = pd.read_parquet("movies.parquet", )
```

```python
df.set_index("ImdbId")
pd.set_option("display.max_colwidth", 10)
df.shape
```

```python
df["genre"].str.split(",").str[0].nunique()
```

```python
df["hauptgenre"] = df["genre"].str.split(",").str[0]
```

```python
#df = df.rename(columns = lambda x : x.lower())
```

```python
df.head()
```

```python
hauptgenre_grp = df.groupby("hauptgenre")
```

```python
hauptgenre_grp.get_group("Comedy")
```

```python
for genre, hauptgenre in hauptgenre_grp:
    hauptgenre.to_parquet( f"/genres/{genre}.parquet")
```
