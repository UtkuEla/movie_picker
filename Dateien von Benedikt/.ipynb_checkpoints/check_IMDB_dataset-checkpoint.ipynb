## IMDB dataset: NaN-vals percentage

```python
# IMDB dataset: NaN-vals percentage

for i in movies.columns:
    null_rate = movies[i].isna().sum() / len(movies) * 100 
    if null_rate > 0 :
        print(f"{i}'s null rate : {null_rate:.2f}%")
```
