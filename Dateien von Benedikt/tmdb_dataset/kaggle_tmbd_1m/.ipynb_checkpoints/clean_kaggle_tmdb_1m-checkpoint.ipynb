```python
import pandas as pd
import numpy as np
import missingno as msno # Via https://www.kaggle.com/code/sadeghjalalian/imdb-best-movies-analysis-recommender-system
```

```python
movies = pd.read_parquet("../tmdb_api/tmdb_filmdaten_2025-03-01_17-42-18.csv")
movies = movies.replace("", np.nan)
```

```python
print(movies.shape)
print(movies.columns)
#print(movies.info)
```

```python
movies = movies.drop(['status', 'revenue', 'runtime', 'adult', 'backdrop_path', 'budget', 'homepage',
       'original_language', 'original_title', 'spoken_languages'], axis=1)
```

```python
msno.matrix(movies)
```

```python
#nans = (movies["keywords"].isnull() | movies["genres"].isnull() | movies["tagline"].isnull())
nans = movies.isnull().any(axis=1)  # Zeilen mit mindestens einem NaN
```

```python
missing_data = pd.DataFrame(movies.loc[nans, "id"]).set_index("id").rename(columns={"id":"movie_id"})
```

```python
missing_data.to_csv("../../input/tmdb_api/links_1m.csv")
```
