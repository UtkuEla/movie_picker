# Merge tmdb 5K dataset with tmdb api-data 

```python
import pandas as pd
import numpy as np
# Kontrolle: 
import missingno as msno # Via https://www.kaggle.com/code/sadeghjalalian/imdb-best-movies-analysis-recommender-system
```

```python
kaggle_data = pd.read_parquet("../../tmbd_datasets_parquet_ready/tmbd_5000_merged.parquet")
```

```python
tmdb_api_movies = pd.read_csv("../../tmdb_api/Archiv/tmdb_filmdaten_2025-02-26_21-24-38.csv")
```

```python
kaggle_data.shape
tmdb_api_movies.shape
```

```python
print(kaggle_data.columns)
print(tmdb_api_movies.columns)
```

```python
#kaggle_data = kaggle_data.rename( columns = { "tmdb_id" : "movie_id"  } )
#kaggle_data.compare(tmdb_api_movies)
```

```python
datasets = [kaggle_data, tmdb_api_movies]
concat = pd.concat(datasets)
```

```python
concat.shape
```

```python
concat = concat.drop(["production_companies", "crew", "production_countries", "spoken_languages"] , axis=1)
concat = concat.drop([614, 547,421])
```

```python
def drop_str(x):
    return pd.to_numeric(x, errors="coerce")
concat["movie_id"] = concat["movie_id"].apply (lambda x : drop_str(x))
concat = concat.dropna(subset = "movie_id")
```

```python
concat["release_date"] = pd.to_datetime(concat["release_date"], errors="coerce")
concat["movie_id"] = pd.to_numeric(concat["movie_id"], errors="coerce")  # Strings -> NaN
concat["movie_id"] = concat["movie_id"].astype(int)  # Jetzt sicher zu int umwandeln
```

```python
concat.to_parquet("../tmbd_exports/tmbd_5000_api-call_merged.parquet")
#concat.to_csv("../tmbd_exports/tmbd_5000_api-call_merged.csv")
```
